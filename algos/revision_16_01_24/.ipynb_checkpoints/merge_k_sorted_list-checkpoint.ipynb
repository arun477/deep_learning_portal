```python
# n: n*k
# s: m+n
# pairwise merging

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def mergeKLists(self, lists):
        def merge(l1, l2):
            if l1 is None:
                return l2
            if l2 is None:
                return l1
            if l1.val <= l2.val:
                l1.next = merge(l1.next, l2)
                return l1
            else:
                l2.next = merge(l2.next, l1)
                return l2
        
        if not lists:
            return None
        head = lists.pop(0)
        while lists:
            head = merge(lists[0], head)
            lists.pop(0)
        return head     
```

```python
# n:
# s:
# divide and conquer approach

class Solution:
    def mergeKLists(self, lists):
        if len(lists)==0:
            return None
            
        def merge(l1, l2):
            if l1 is None:
                return l2
            if l2 is None:
                return l1
            if l1.val <= l2.val:
                l1.next = merge(l1.next, l2)
                return l1
            else:
                l2.next = merge(l2.next, l1)
                return l2
        
        def merge_list(lists, start, end):
            if start==end:
                return lists[start]
            mid = (start+end)//2
            left = merge_list(lists, start, mid)
            right = merge_list(lists, mid+1, end)
            return merge(left, right)
        
        return merge_list(lists, 0, len(lists)-1)   
```

```python
# n:nlogk
# s:l
# heapq minimum priority queue approach

import heapq

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        heap = []
        for i,lst in enumerate(lists):
            if lst:
                heapq.heappush(heap, (lst.val, i, lst)) # logn

        dummy = ListNode()
        curr = dummy
        while heap:
            val,i,node = heapq.heappop(heap) # logn
            curr.next = node
            curr = node
            if node.next:
                heapq.heappush(heap, (node.next.val, i, node.next))
        return dummy.next
        
```
