```python
# t: n^2
# s: 1


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeTwoList(list1, list2):
    if list1 and not list2:
        return list1
    
    if list2 and not list1:
        return list2

    if not(list1 and list2):
        return None
    
    left, right = list1, list2
    head = ListNode
    dummy = head
    
    while left and right:
        if left.val <= right.val:
            temp = left.next
            left.next = None
            dummy.next = left
            dummy = dummy.next
            left = temp
        else:
            temp = right.next
            right.next = None
            dummy.next = right
            dummy = dummy.next
            right = temp
    
    if left:
        dummy.next = left
    if right:
        dummy.next = right
    
    return head.next
            

class Solution:
    def mergeKLists(self, lists: [[ListNode]]) -> [ListNode]:
        if not lists:
            return None
        
        n = len(lists)
        head =  lists[0]
        for i in range(1, n):
            head = mergeTwoList(head, lists[i])
        
        return head
```

```python
# t: nlogk, k: num of linked lists
# s: k

import heapq

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: [[ListNode]]) -> [ListNode]:
        dummy =  ListNode()
        current = dummy
        heap = []
        
        for i in range(len(lists)):
            if lists[i]:
                heapq.heappush(heap, (lists[i].val, i))
                lists[i] = lists[i].next
        
        while heap:
            val, i = heapq.heappop(heap)
            current.next = ListNode(val)
            current = current.next
            
            if lists[i]:
                heapq.heappush(heap, (lists[i].val, i))
                lists[i] = lists[i].next
        
        return dummy.next
```
