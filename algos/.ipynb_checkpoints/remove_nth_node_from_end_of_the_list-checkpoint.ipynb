Given the head of a linked list, remove the nth node from the end of the list and return its head.

 

Example 1:


Input: head = [1,2,3,4,5], n = 2
Output: [1,2,3,5]
Example 2:

Input: head = [1], n = 1
Output: []
Example 3:

Input: head = [1,2], n = 1
Output: [1]
 

Constraints:

The number of nodes in the list is sz.
1 <= sz <= 30
0 <= Node.val <= 100
1 <= n <= sz
 

Follow up: Could you do this in one pass?

```python
# n: n
# s: n
# 
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        h = {}
        curr = head
        i=1
        while curr:
            h[i] = curr
            curr = curr.next
            i +=1
        if len(h)==1:
            return None
        else:
            if len(h)-n <= 0:
                h[len(h)-n+1].next = None
                return h[len(h)-n+2]
            if len(h)-n in h:
                h[len(h)-n].next = None
                if len(h)-n+2 in h:
                    h[len(h)-n].next = h[len(h)-n+2] 
        return head   
```
