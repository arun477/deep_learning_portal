```python
import numpy as np
import torch
from torch import tensor
import math
```

```python
n_clusters = 6
n_samples = 250
centroids = torch.randn((n_clusters, 2))*70-35
```

```python
from torch.distributions.multivariate_normal import MultivariateNormal
def multsample(m):
    return MultivariateNormal(m, torch.diag(tensor([5., 5.]))).sample((n_samples,))
```

```python
samples = [multsample(c) for c in centroids]
data = torch.cat(samples)
data.shape
```

```python
import matplotlib.pyplot as plt

def plot_data(centroids, data):
    _,ax = plt.subplots()
    for i,centroid in enumerate(centroids):
        samples = data[i*n_samples:(i+1)*n_samples]
        ax.scatter(samples[:,0], samples[:,1], s=1)
        ax.plot(*centroid, marker='x', markersize=10, color='k', mew=5)
```

```python
plot_data(centroids, data)
```

```python
midp = data.mean(0)
plot_data([midp]*n_clusters, data)
```

```python
def gauss_kernel(d, bw):
    return torch.exp(-0.5*((d/bw))**2) / (bw*math.sqrt(2*math.pi))
```

```python
def meanshift_update(X, h):
    for i,x in enumerate(X):
        dists = ((x-X)**2).sum(1).sqrt()
        weights = gauss_kernel(dists, h)
        X[i] = (X*weights[:,None]).sum(0)/weights.sum()
    return X
```

```python
def meanshift(data, n_update=5, h=2.5):
    X = data.clone()
    for _ in range(n_update):
        meanshift_update(X, h)
    return X
```

```python
%time X1 = meanshift(data)
```

```python
plot_data(centroids+10, X1)
```

```python
from itertools import product

def meanshift_pp_update(X, h, dim):
    nearby_grids = list(product([-1, 0, 1], repeat=dim))
    C,S = {},{}
    for _,x in enumerate(X):
        idx = (x/h).floor()
        idx = tuple(idx.tolist())
        if idx not in C:
            C[idx] = 0
            S[idx] = torch.zeros_like(X[0])
        C[idx] += 1
        S[idx] = S[idx] + tensor(x)
    for i,x in enumerate(X):
        sums,counts = torch.zeros_like(x),0
        for ngrd in nearby_grids:
            idx = tuple(((x/h).floor() + tensor(ngrd)).tolist())
            idx = tuple(idx) 
            if idx in S:
                sums += S[idx]
                counts += C[idx]
        X[i] = sums/counts 
    return C
```

```python
def meanshift_pp(data, h=2.5, n_updates=5):
    X = data.clone().detach() 
    for _ in range(n_updates):
        C = meanshift_pp_update(X, h, dim=X.shape[1])
    return X, C
```

```python
%time X2,C = meanshift_pp(data, n_updates=5)
```

```python
plot_data(centroids+15, X2)
plt.scatter(tensor(list(C.keys()))[:,0], tensor(list(C.keys()))[:,1])
```
