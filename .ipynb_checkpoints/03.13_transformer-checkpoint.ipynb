```python
import numpy as np
import torch
import torch.nn.functional as F
import math, time
import matplotlib.pyplot as plt
import seaborn as sns

# plot look context, seaborn set look and the feel of matplotlib plot under the hood
sns.set_context(context='talk')

%matplotlib inline
```

```python
# checking how different context setting of seaborn changes the 
# matplotlib style context.

contexts = ['paper', 'notebook', 'talk', 'poster']
data = sns.load_dataset('iris')
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

# with direct matplotlib scatter
_, axs = plt.subplots(1, 4, figsize=(20, 3))
for cntx, ax in zip(contexts, axs): 
    sns.set_context(context=cntx)
    ax.scatter(data.sepal_length.tolist(), data.sepal_width.tolist(), c=data['species'].map(colors));
    ax.set_title(f'context:{cntx}', fontsize=12)
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.tick_params(axis='x', labelsize=12)  
    ax.tick_params(axis='y', labelsize=12) 
    

# with seaborn scatter
_, axs = plt.subplots(1, 4, figsize=(20, 3))
for cntx, ax in zip(contexts, axs): 
    sns.set_context(context=cntx)
    sns.scatterplot(x='sepal_length', y='sepal_width', data=data, ax=ax, hue='species')
    ax.set_title(f'context:{cntx}', fontsize=12)
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.tick_params(axis='x', labelsize=12)  
    ax.tick_params(axis='y', labelsize=12) 
    ax.legend([])
plt.show();
```
