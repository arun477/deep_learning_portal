```python
batch_size = 32
block_size = 8 # context length
max_iters = 3000
eval_interval = 300
learning_rate = 1e-2
eval_iters = 200
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# note: it won't persist across cells in notebooks
torch.manual_seed(1337);
```

```python
# data from: https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt
with open('./data/tinyshakespeare.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
# encode characters
chars = sorted(list(set(text)))
vocab_size = len(chars)
stoi = {s:i for i,s in enumerate(chars)}
itos = {i:s for s,i in stoi.items()}
encode = lambda w: list(stoi[ch] for ch in w)
decode = lambda idx: "".join(list(itos[i] for i in idx))
```

```python
data = torch.tensor(encode(text), dtype=torch.long)

# do train/test split, 90%-10% train-test split of total size
n = int(0.9*len(data))
train_data = data[:n]
val_data = data[n:]
```

```python
def get_batch(split):
    data = train_data if split=='train' else val_data
    
    # len(data)-block_size (to avoid overflow based on context)
    idx =  torch.randint(len(data)-block_size, (batch_size,))
    
    # chunkc of chars. size is based on block_size (or context length).
    x = torch.stack([data[i:i+block_size] for i in idx])
    
    # shift chars 1 step right, for char level prediction label.
    y = torch.stack([data[i+1:i+block_size+1] for i in idx])
    x, y = x.to(device), y.to(device)
    
    return x, y
```

```python
@torch.no_grad()
def estimate_loss(model):
    model.eval()
    o = {}
    for split in ['train', 'val']:
        losses = torch.zeros(eval_iters)
        for k in range(eval_iters):
            X, Y = get_batch(split)
            logits, loss = model(X, Y)
            losses[k] = loss.item()
        o[split] = losses.mean()
    model.train()
    return o    
```

```python
class BigramLanguageModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        
        # single embeding layer which act as direct next character logits lookup table.
        # first dimension is the input char, second dimension will give logits 
        # for all possible next char in vocab.
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)
        
    def forward(self, x, targets=None):
        
        logits = self.token_embedding_table(x)
        
        if targets is None:
            loss = None
        else:
            # B: batch size.
            # T: block size.
            # C: next char logits size based of each item in T.
            B, T, C = logits.shape
            
            # shape changes: for next char logits (vocab size), with single correct target index
            # F.cross_entropy handle this shape, to actualy calculate the cross entropy
            logits = logits.view(B*T, C)
            targets = targets.view(B*T)
            
            loss = F.cross_entropy(logits, targets)
         
        return logits, loss
    
    @torch.no_grad()
    def generate(self, idx, max_token):
        # idx: shape B,T, start with B=1 and T=1
        
        for _ in range(max_token):
            
            logits, loss = self(idx)
            
            # logits shape will be B,T,C
            # each loop next character index will be added back to the idx
            # but we don't need previous character prediction only new character prediction
            # based on last token.
            # so it becomes B,C
            logits = logits[:, -1 ,:]
            
            # calc probs from logits
            probs = F.softmax(logits, dim=-1)
            
            # sample next token using multinomial distibution istead of picking high one.
            idx_next = torch.multinomial(probs, num_samples=1) # shape will be B,1
            
            
            
                
```

```python
model = BigramLanguageModel(vocab_size)
m = model.to(device)
opt = torch.optim.AdamW(model.parameters(), lr=learning_rate)
```

```python
for itr in range(max_iters):
    if itr%eval_interval==0:
        o = estimate_loss(m)
        print(f"train loss: {o['train']:.4f}, val loss: {o['val']:.4f}")
        
    xb, yb = get_batch('train')
    
    logits, loss = m(xb, yb)
    
    opt.zero_grad(set_to_none=True)
    
    loss.backward()
    
    opt.step()
```
