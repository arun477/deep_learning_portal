```python
import torch
```

```python
chars = sorted(list(set("".join(words))))
stoi= {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
```

```python
N = torch.zeros((len(stoi), len(stoi)), dtype=torch.int32)
```

```python
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1,ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues');

for i in range(len(stoi)):
    for j in range(len(stoi)):
        chrstr = itos[i]+itos[j]
        plt.text(j, i, chrstr, ha='center', va='bottom', color='gray')
        plt.text(j, i, N[i,j].item(), ha='center', va='top', color='gray')
plt.axis('off');
```

```python
p = N[0].float()
p = p/p.sum()
p
```

```python
ix = torch.multinomial(p, num_samples=1, replacement=True).item()
itos[ix]
```

```python
P = (N+1).float()
P = P/P.sum(1, keepdim=True)
```

```python
for _ in range(5):
    ix = 0
    outs = []
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True).item()
        outs.append(itos[ix])
        if ix==0:
            break
    print(''.join(outs))
```

```python
log_likelihood = 0
n = 0
for word in words:
    chrs = ['.'] + list(word) + ['.']
    for ch1,ch2 in zip(chs, chs[1:]):
        ix1,ix2 = stoi[ch1],stoi[ch2]
        prob = P[ix1,ix2]
        log_prob = torch.log(prob)
        log_likelihood += log_prob
        n +=1

nnl = -log_likelihood
avg_nnl = nnl/n
print('log likelihood', log_likelihood)
print('nll', nnl)
print('avg nnl', avg_nnl)
```

```python
xs,ys = [],[]
for word in words:
    chrs = ['.'] + list(word) + ['.']
    for ch1,ch2 in zip(chrs, chrs[1:]):
        ix1,ix2 = stoi[ch1],stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)
xs = torch.tensor(xs)
ys = torch.tensor(ys)
n_samples = xs.nelement()
print('n_samples', n_samples)
```

```python
W = torch.randn(xenc.shape[1], xenc.shape[1], requires_grad=True)
```

```python
for i in range(5):
    xenc = F.one_hot(xs, len(itos)).float()
    logits = xenc@W
    counts = logits.exp()
    probs = counts/counts.sum(1, keepdim=True)
    loss = -probs[torch.arange(len(xs)), ys].log().mean() + 0.01*(W**2).mean()
    print(loss.item())
    
    W.grad = None
    loss.backward()
    
    W.data += -50*W.grad
```

```python
for i in range(5):
    out = []
    ix = 0
    while True:
        xenc = F.one_hot(torch.tensor([ix]), len(itos)).float()
        logits = xenc@W
        counts = logits.exp()
        p = counts/counts.sum(1, keepdim=True)
        
        ix = torch.multinomial(p, num_samples=1, replacement=True).item()
        out.append(itos[ix])
        if ix==0:
            break
    print(''.join(out))
```
