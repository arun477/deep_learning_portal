```
from datasets import load_dataset
```

```
minds = load_dataset("PolyAI/minds14", name='en-AU', split='train')
```

```
from transformers import pipeline
import copy
```

```
cls = pipeline('audio-classification', model="anton-l/xtreme_s_xlsr_300m_minds14")
```

```
preds = cls(minds[0]['audio'])
```

```
preds
```

```
id2label = minds.features['intent_class'].int2str
id2label(minds[0]['intent_class'])
```

```
speech_commands = load_dataset("speech_commands", "v0.02", split='validation', streaming=True)
```

```
# examples = list(speech_commands.take(100))
# example = examples[0]
example = next(iter(speech_commands))
example
```
