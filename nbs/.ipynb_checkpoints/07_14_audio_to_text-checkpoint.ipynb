```
from datasets import load_dataset
```

```
dataset = load_dataset('facebook/voxpopuli', name='en', streaming=True, split='train')
```

```
examples = dataset.take(5)
examples = list(examples)
```

```
example = examples[2]
example
```

```
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```
D = librosa.stft(example['audio']['array'])
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```
plt.figure().set_figwidth(15);
librosa.display.specshow(S_db, x_axis='time', y_axis='hz');
plt.colorbar();
```

```
S = librosa.feature.melspectrogram(y=example['audio']['array'], sr=example['audio']['sampling_rate'], n_mels=128, fmax=8000)
S_db = librosa.power_to_db(S, ref=np.max)
S_db.shape
```

```
plt.figure().set_figwidth(15);
librosa.display.specshow(S_db, x_axis='time', y_axis='hz');
plt.colorbar();
```
