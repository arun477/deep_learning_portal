```python
import numpy as np
import scipy
import scipy.io.wavfile
import librosa
import matplotlib.pyplot as plt
import IPython as ipy
```

```python
sr, signal = scipy.io.wavfile.read('../datasets/OSR_us_000_0010_8k.wav')
signal = signal / np.max(np.abs(signal))
sr,signal.shape
```

```python
signal2, sr2 = librosa.load('../datasets/OSR_us_000_0010_8k.wav', sr=8000)
sr2, signal2.shape
```

```python
np.allclose(signal2, signal, rtol=0.5)
```

```python
print(signal.min(), signal.max())
print(signal2.min(), signal2.max())
```

```python
# lets take first 3.5 sec of the audio signal

signal = signal[0: int(3.5*sr)]
signal.shape
```

```python
plt.figure(figsize=(15, 5))
librosa.display.waveshow(signal, sr=sr);
```

```python
ipy.display.Audio(signal, rate=sr)
```

The pre-emphasis filter can be applied to a signal x
 using the first order filter in the following equation:

y(t)=x(t)x(t1)

```python
alpha = 0.97
emphasized_signal = signal[1:]-alpha*signal[:-1]
```

```python
plt.figure(figsize=(15, 5))
librosa.display.waveshow(emphasized_signal, sr=sr);
```

```python
ipy.display.Audio(emphasized_signal, rate=sr)
```
